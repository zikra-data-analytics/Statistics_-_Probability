```python
data  = [50, 70 , 60 , 90 , 35]
# average = (50 + 70 + 60 + 90 + 35)/5 = 61
print(np.mean(data))
# var = (50 - 61)^2 + (70 - 61)^2+ (60 - 61)^2 + (90 - 61)^2 + (35 - 61)^2 / 5
# var = (-11)^2 + (9)^2 + (-1)^2 + (29)^2 + (-26)^2 /5
# var = 1720 /5 = 344
```

```python
# HIGH VARIANCE -> values are highly spread / inconsistent
# LOW VARIANCE -> values are less spread / tightly packed
```

```python
# example -
company_1 = [40000 , 42000 , 39000 , 41000 , 43000] # everyone earns roughly the same
company_2 = [10000 , 20000 , 40000 , 80000 , 100000] # huge salary inequality
print('variance for company 1- ',np.var(company_1))
print('variance for company 2- ',np.var(company_2))
print('mean of company 1-',np.mean(company_1))
print('mean of company 2-',np.mean(company_2))
# COMPANY 1 - has low variance
# COMPANY 2- high variance
```

```python
import numpy as np
# 1.Range
Marks = [70, 85, 90, 60, 75]
Salaries = [20000, 25000, 30000, 22000, 28000]
Temperature = [32, 35, 33, 36, 34]

# Range = Max - Min
print("Marks Range:", max(Marks) - min(Marks))
print("Salary Range:", max(Salaries) - min(Salaries))
print("Temperature Range:", max(Temperature) - min(Temperature))
```

```python
# Example: Student Marks
# 2. Variance
import numpy as np
import statistics as stats

Marks = [70, 85, 90, 60, 75]
Salaries = [20000, 25000, 30000, 22000, 28000]
Temperature = [32, 35, 33, 36, 34]

# Using numpy
print("Marks Variance:", np.var(Marks))
print("Salary Variance:", np.var(Salaries))
print("Temperature Variance:", np.var(Temperature))

# Using statistics
print("Marks Variance (stats):", stats.variance(Marks))
```

```python
# 3. Standard Deviation

import numpy as np
import statistics as stats

Marks = [70, 85, 90, 60, 75]
Salaries = [20000, 25000, 30000, 22000, 28000]
Temperature = [32, 35, 33, 36, 34]

# Using numpy
print("Marks SD:", np.std(Marks))
print("Salary SD:", np.std(Salaries))
print("Temperature SD:", np.std(Temperature))

# Using statistics
print("Marks SD (stats):", stats.stdev(Marks))
```
